### 1.数据预处理

```python
import numpy as np
import pandas as pd
pd.set_option('display.max_columns',None)
```

```python
df = pd.read_excel('materical_group.xlsx')
df
```

```python
risk_columns = ['risk' + str(x) for x in range(1, 25)]
money_columns = ['money' + str(x) for x in range(1, 10)]
risk_columns
```

### 创造规律性，计算risk_columns总分，进行分类，创作规律

```python
df['risk_score'] = 0
for column in risk_columns:
    df['risk_score'] += df[column]

# 创建 money_score 需要初始化一个值 0 或是 defaultdict()
df['money_score'] = 0
for column in money_columns:
    df['money_score'] += df[column]
df
```

```python
# 计算risk_score和money_risk的中位数，大于中位数变为一个值，小于中位数变为一个字，方便LR学出来
risk_median = df['risk_score'].quantile(q=[0.5])[0.5] # 0.5    11942.0
# risk_median = df['risk_score'].median()
print(risk_median)
money_median = df['money_score'].median()
print(money_median)
```

```python
# 项目中反复使用：辅助LR找到规律
'''模拟人工标记'''
for index, row in df.iterrows():
    if row['risk_score'] > risk_median:
        # index帮助锁定 risk_labe的值
        df.loc[index, 'risk_label'] = np.random.choice([0, 1], p=[0.01, 0.99]) 
    else:
        df.loc[index, 'risk_label'] = np.random.choice([0, 1], p=[0.99, 0.01])
df
# 在(0, 1]之间进行选取，p 规定选取a中每个元素的概率，默认为概率相同

for index, row in df.iterrows():
    if row['money_score'] > risk_median:
        # 使用 .loc 锁定对这一行进行操作，否则默认对列进行操作
        df.loc[index, 'money_label'] = np.random.choice([0, 1], p=[0.01, 0.99])
    else:
        df.loc[index, 'money_label'] = np.random.choice([0, 1], p=[0.99, 0.01])
df
```

```python
df.to_excel('./materical_group2.xlsx','wb')
```
