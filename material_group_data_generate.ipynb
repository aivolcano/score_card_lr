### 1.生成材料组数据

```python
import pandas as pd
import numpy as np
pd.set_option('display.max_columns',None)
```

```python
num = 400
columns_name = ['material_group_code','material_group_name']
df = pd.DataFrame(columns=columns_name)
df['material_group_code'] = range(1, num+1) # 左闭右开的区间
df['material_group_name'] = ['材料组'+ str(x) for x in range(1, num+1)]
df
```

### 2.生成24个字段

```python
# risk 24个字段，money 9个字段
for i in range(24):
    temp = 'risk' + str(i+1) # 从risk1开始不是risk0
    # randint (正太分布)
    columns_name.append(temp)
    # randint(start, end, num)
    df[temp] = np.random.randint(1, 1000, num) # [1, 1000)范围内的值， 共num个
df
# randint符合正态分布 均值为0 方差为1 
```

```python
# money字段
for i in range(9):
    temp = 'money' + str(i+1)
    columns_name.append(temp)
    df[temp] = np.random.randint(1, 1000, num)
df
```

```python
# 人工打label  0-1
df['risk_label'] = np.random.randint(0, 2, num)
df['money_label'] = np.random.randint(0, 2, num)
```

### 赋予数据规律性

```python
df.head()
```

```python
risk_columns = ['risk' + str(x) for x in range(1, 25)]
money_columns = ['money' + str(x) for x in range(1, 10)]
print(risk_columns)
```

### 创造规律性
计算risk_columns总分，进行分类，创作规律

```python
df['risk_score'] = 0
for column in risk_columns:
    df['risk_score'] += df[column]

# 创建 money_score 需要初始化一个值 0 或是 defaultdict()
df['money_score'] = 0
for column in money_columns:
    df['money_score'] += df[column]
df.head()
```

```python
# 计算risk_score和money_risk的中位数，大于中位数变为一个值，小于中位数变为一个字，方便LR学出来
risk_median = df['risk_score'].quantile(q=[0.5])[0.5]
print(risk_median)
money_median = df['money_score'].median()
print(money_median)
```

```python
# 辅助LR找规律
'''模拟人工标记: risk'''
for index, row in df.iterrows():
    if row['risk_score'] > risk_median:
        # index帮助锁定 risk_labe的值
        df.loc[index, 'risk_label'] = np.random.choice([0, 1], p=[0.01, 0.99])
    else:
        df.loc[index, 'risk_label'] = np.random.choice([0, 1], p=[0.99, 0.01])

'''模拟人工标记: money'''
for index, row in df.iterrows():
    if row['money_score'] > risk_median:
        # 使用 .loc 锁定对这一行进行操作，否则默认对列进行操作
        df.loc[index, 'money_label'] = np.random.choice([0, 1], p=[0.01, 0.99])
    else:
        df.loc[index, 'money_label'] = np.random.choice([0, 1], p=[0.99, 0.01])
```
